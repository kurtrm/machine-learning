```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
NOAA = pd.read_csv("NOAA_data.csv")
```

```python
#  Need to format the Date to match the fremont dates
NOAA['DATE'] = pd.to_datetime(NOAA['DATE'].astype(str))
```

```python
NOAA.rename(columns={'DATE': 'Date'}, inplace=True)
```

```python
NOAA.set_index('Date')
```

```python
NOAA_date_and_prcp = NOAA[['Date', 'PRCP']]
```

```python
NOAA_date_and_prcp.set_index('Date')
```

```python
fremont_bike_count = pd.read_csv("fremont-bike-count.csv")
```

```python
number_of_nulls = fremont_bike_count.isnull().sum()  # Just for my info
```

```python
#  I chose to fill the null fields with the median value
```

```python
filled_fremont_bike_count = fremont_bike_count.fillna(fremont_bike_count[['Fremont Bridge East Sidewalk',
                                              'Fremont Bridge West Sidewalk']].median())
```

```python
nulls_in_filled = filled_fremont_bike_count.isnull().sum()  # For my info
```

```python
#  Now I need to aggregate the hours into days for the bridge data
filled_fremont_bike_count['Date'] = pd.to_datetime(filled_fremont_bike_count['Date'])
```

```python
aggregated_bike_count = (filled_fremont_bike_count
.set_index(pd.DatetimeIndex(filled_fremont_bike_count['Date']))
.groupby(pd.TimeGrouper('D'))
.aggregate(np.sum)
.reset_index())
```

```python
bridge_and_precip_daily = aggregated_bike_count.merge(NOAA_date_and_prcp) # This took me forever to figure out
```

```python
#  Chose to aggregate monthly as this would make a graph a little easier to manage
bridge_and_precip_monthly = (bridge_and_precip_daily
.set_index(pd.DatetimeIndex(bridge_and_precip_daily['Date']))
.groupby(pd.TimeGrouper('M'))
.aggregate(np.sum)
.reset_index())
```

```python
bridge_and_precip_monthly.columns = ['Date', 'West Sidewalk', 'East Sidewalk', 'Rain']
```

```python
xlabel = bridge_and_precip_monthly['Date'].dt.strftime('%b \'%y')
title = 'Fremont Bridge Bicycle Crossings By Month'
bar_chart = (bridge_and_precip_monthly[['Date',
                                       'East Sidewalk',
                                       'West Sidewalk']]
            .plot(kind='bar',
                  x=xlabel,
                  rot=90,
                  title=title,
                  figsize=(8, 4)))
```

```python
# plt.xlabel("Month")
# plt.ylabel("# Bikes Crossing Fremont Bridge Northbound/Southbound")
bar_chart.set_xlabel('Month and Year')
bar_chart.set_ylabel("# Bike Crossings")
```

```python
line_chart = bridge_and_precip_monthly[['Date', 'Rain']].plot(linestyle='-',
                                               marker='o',
                                               secondary_y=True,
                                               rot=90,
                                               ax=bar_chart,
                                               color='#009999',
                                               label='Inches')
```

```python
line_chart.axhline(c='black', lw=.2)
line_chart.set_ylim(-15, 15)
line_chart.set_ylabel("inches")
```

```python
plt.show()
```
